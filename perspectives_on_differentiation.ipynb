# 미분은 '기울기 함수'를 도출하는 과정이다

보통 "미분은 변화율이다"라고 배운다. 하지만 이 설명은 너무 단순하다.  
미분을 더 본질적으로 바라보면, 그것은 **입력이 변화할 때 출력이 어떤 규칙으로 반응하는지를 나타낸 함수**다.

---

## 1. 함수란 무엇인가?

함수란 입력과 출력의 관계를 정리한 규칙이다.  
예를 들어 $ y = f(x) = x^2$ 라는 함수는 "x를 넣었을 때 y가 어떻게 나오는가"를 설명하는 법칙이다.

하지만 우리는 단순히 어떤 x에 대한 y만 알고 싶은 것이 아니라,  
**x가 조금 변했을 때 y는 어떻게 변하는가**를 알고 싶어지는 순간이 있다.

그때 등장하는 개념이 **미분**이다.

---

## 2. 미분은 '변화 규칙'을 도출하는 연산이다

함수 $f(x)$가 있을 때, 특정한 기준점 $a$에서 입력값을 아주 조금 $\Delta x$만큼 바꾸면 출력값도 이에 따라 바뀐다:

- $x: a \rightarrow a + \Delta x$
- $y: f(a) \rightarrow f(a + \Delta x)$

이때의 변화율은 다음과 같이 표현할 수 있다:

$$
\frac{\Delta y}{\Delta x} = \frac{f(a + \Delta x) - f(a)}{\Delta x}
$$

$  \Delta x $를 작게 줄여가며 계산한 변화율이 어떤 일정한 경향을 가진다면, 우리는 이 변화율 자체를 또 하나의 함수로 정의할 수 있다.   
그것이 바로 **도함수**이며, 우리는 이 과정을 **미분한다**고 표현한다.

> 이때 $\Delta x$를 무한히 작게 만든 순간 변화율, 즉 도함수는 다음과 같이 정의된다:
$$
f'(a) = \lim_{\Delta x \to 0} \frac{f(a + \Delta x) - f(a)}{\Delta x}
$$

---

## 3. "~에 대해 미분한다"는 의미

"$ x$에 대해 미분한다"는 말은,  
**x를 바꿨을 때 y가 어떻게 반응하는지를 본다**는 의미다.

즉, 독립 변수인 x를 살짝 움직여보면서 종속 변수인 y가 어떻게 반응하는지를 수학적으로 정리하는 것이다.  
이 관점에서 보면 미분은 단순한 계산이 아니라, **변화의 방향성과 민감도를 이해하는 사고 방식**이다.

---

## 4. 미분은 '기울기 지도'를 만들어준다

원래 함수 $ f(x)$는 x를 넣으면 y를 출력한다.  
반면에 도함수 $f'(x)$는 x를 넣으면 "그 지점에서의 기울기"를 출력한다.

즉, 미분은 다음과 같은 과정을 통해 기울기만 따로 모은 새로운 함수를 만든다:

- 입력 x를 조금 바꿨을 때의 변화율을 측정한다
- 그 결과값을 또 다른 함수로 표현한다

이렇게 얻은 함수는 마치 **기울기 지도(gradient map)** 처럼 작동한다.

---

## 5. 미분이 중요한 이유

- 어떤 시스템이 얼마나 민감하게 반응하는지 측정할 수 있다
- 변화의 방향과 속도를 알 수 있다
- 인공지능에서는 이 기울기 함수를 따라가며 학습한다 (예: Gradient Descent)

---

## 6. 정리

미분은 x가 변할 때 y가 어떻게 반응하는지를 분석해  
그 **변화율을 또 하나의 함수로 표현하는 과정**이다.

이를 통해 우리는 수학적으로 **변화의 규칙**을 구조화하고,  
현상을 더 깊이 이해할 수 있게 된다.
